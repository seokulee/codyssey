# 수행단계
- 장르 탐색
    - 영화의 장르별 분포 (Bar chart) 및 장르별 평균 평점을 확인하라.

```python
import matplotlib as mpl

mpl.rcParams['font.family'] = 'AppleGothic'
mpl.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


genres_df = pd.read_csv("data/kmrd/genres.csv")
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

sns.countplot(data=genres_df, x="genre", ax=ax, order=genres_df["genre"].value_counts().index)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 장르별 평균 평점

ratings_df = pd.read_csv("data/kmrd/rates.csv")
movies_df = pd.read_csv("data/kmrd/movies.txt", sep="\t", header=None, names=["movie", "title", "title_eng", "year", "grade"])
genres_df = pd.read_csv("data/kmrd/genres.csv")

ratings_df["movie"] = ratings_df["movie"].astype(str)
genres_df["movie"] = genres_df["movie"].astype(str)

movie_ratings_df = pd.merge(ratings_df, movies_df, on="movie")
movie_ratings_df
```

```python
movie_ratings_df = pd.merge(movie_ratings_df, genres_df, on="movie")
movie_ratings_df
```

```python
movie_ratings_df.groupby("genre")["rate"].mean().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(10, 5))

sns.ba`rplot(data=movie_ratings_df, x="genre", y="rate", ax=ax)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
